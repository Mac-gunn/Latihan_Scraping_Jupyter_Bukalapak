```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver')
```

```python
products=[]
prices=[]
image=[]
driver.get('https://www.bukalapak.com/products?from=omnisearch&from_keyword_history=false&search%5Bkeywords%5D=kevlar&search_source=omnisearch_category&source=navbar')
```

```python
content = driver.page_source
```

```python
soup = BeautifulSoup(content)
for link in soup.find_all('div',class_="product-card"):
    print(link)
```

```python
for link in soup.find_all('div',class_="product-card"):
    dataImg = link.find('img', class_="product-media__img")
    name = link.find('a', class_="product__name line-clamp--2 js-tracker-product-link qa-list")
    price = link.find('span', class_="amount positive")
    #Memasukan Data
    products.append(name.get_text())
    prices.append(price.get_text())
    image.append(dataImg['src'])
```

```python
df = pd.DataFrame({'Productname':products," Price":prices,"Image":image})
df.to_excel('products.xls', index=False, encoding='utf-8')
```
